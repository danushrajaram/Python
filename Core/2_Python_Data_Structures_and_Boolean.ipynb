### Boolean Variables

Boolean values are the two constant objects False and True.

They are used to represent truth values (other values can also be considered
false or true).

In numeric contexts (for example, when used as the argument to an
arithmetic operator), they behave like the integers 0 and 1, respectively.

The built-in function bool() can be used to cast any value to a Boolean,
if the value can be interpreted as a truth value

They are written as False and True, respectively.

## Dictionaries

A dictionary is a collection which is unordered, changeable and indexed. In Python dictionaries are written with curly brackets, and they have keys and values.

A dictionary is a collection that stores information in pairs — one part called a key and the other a value.

You use the key to find the value that goes with it, like looking up a word to find its meaning. Each key is unique, and every value is linked to one key.

Dictionaries are used to organize data so it’s easy to search, group, or update. In simple terms, it’s a structure for matching “this” with “that.”

## Tuples

A tuple is an ordered collection of items — kind of like a list — but it can’t be changed once created.

You can think of it as a fixed package of data. Each item has a specific position, and the order stays the same.

Tuples are often used when you want to group information that naturally belongs together but shouldn’t be modified — for example, coordinates, dates, or database records.

They’re reliable, lightweight, and slightly faster than lists because of that immutability.

```python
# ==============================================================
# 🐍 PYTHON MINI PROJECT – "All-in-One Interactive Concepts Demo"
# Author: Danush (you 😉)
# Description: A single interactive Python script that demonstrates
#              core Python concepts in one place.
# ==============================================================

print("===== WELCOME TO PYTHON CONCEPTS DEMO =====")

# --------------------------------------------------------------
# 1️⃣ VARIABLES, DATA TYPES, AND TYPE CASTING
# --------------------------------------------------------------
print("\n--- Variables & Data Types ---")
a = input("Enter a number: ")
b = input("Enter another number: ")
a = int(a)
b = float(b)

print(f"a = {a} (type: {type(a).__name__})")
print(f"b = {b} (type: {type(b).__name__})")

# Type casting
sum_result = a + int(b)
print("Sum after type casting:", sum_result)

# --------------------------------------------------------------
# 2️⃣ ARITHMETIC & ASSIGNMENT OPERATORS
# --------------------------------------------------------------
print("\n--- Arithmetic & Assignment Operators ---")
print(f"Addition: {a + b}")
print(f"Subtraction: {a - b}")
print(f"Multiplication: {a * b}")
print(f"Division: {a / b}")
print(f"Modulus: {a % b}")

a += 5
print("After a += 5 →", a)

# --------------------------------------------------------------
# 3️⃣ BOOLEAN, LOGICAL, & COMPARISON OPERATORS
# --------------------------------------------------------------
print("\n--- Boolean, Logical, and Comparison Operators ---")
bool1 = bool(int(input("Enter 1 for True or 0 for False: ")))
bool2 = bool(int(input("Enter another (1 for True, 0 for False): ")))

print(f"{bool1} AND {bool2} → {bool1 and bool2}")
print(f"{bool1} OR {bool2} → {bool1 or bool2}")
print(f"NOT {bool1} → {not bool1}")

x = int(input("Enter a number to compare: "))
y = int(input("Enter another number to compare: "))

print(f"{x} == {y} → {x == y}")
print(f"{x} > {y} → {x > y}")
print(f"{x} <= {y} → {x <= y}")

# --------------------------------------------------------------
# 4️⃣ STRINGS
# --------------------------------------------------------------
print("\n--- String Operations ---")
text = input("Enter a word or sentence: ")
print("Uppercase:", text.upper())
print("Lowercase:", text.lower())
print("Reversed:", text[::-1])
print("Word count:", len(text.split()))

# --------------------------------------------------------------
# 5️⃣ LISTS
# --------------------------------------------------------------
print("\n--- List Operations ---")
lst = input("Enter list elements separated by spaces: ").split()
print("Original List:", lst)
lst.append(input("Add an element to the end: "))
lst.insert(0, input("Insert an element at the beginning: "))
print("After modifications:", lst)
lst.pop()
print("After pop:", lst)
lst.sort()
print("After sort:", lst)
print("Slice (first 3 elements):", lst[:3])

# --------------------------------------------------------------
# 6️⃣ TUPLES
# --------------------------------------------------------------
print("\n--- Tuple Operations ---")
tup = tuple(input("Enter tuple elements separated by spaces: ").split())
print("Tuple:", tup)
print("Length:", len(tup))
print("First element:", tup[0] if tup else "Empty tuple")

# --------------------------------------------------------------
# 7️⃣ SETS
# --------------------------------------------------------------
print("\n--- Set Operations ---")
set1 = set(input("Enter Set 1 elements: ").split())
set2 = set(input("Enter Set 2 elements: ").split())

print("Union:", set1 | set2)
print("Intersection:", set1 & set2)
print("Difference (Set1 - Set2):", set1 - set2)
print("Symmetric Difference:", set1 ^ set2)

# --------------------------------------------------------------
# 8️⃣ DICTIONARIES
# --------------------------------------------------------------
print("\n--- Dictionary Operations ---")
d = {}
n = int(input("How many key-value pairs? "))
for i in range(n):
    key = input("Enter key: ")
    val = input("Enter value: ")
    d[key] = val
print("Dictionary:", d)

search = input("Enter a key to search: ")
print("Value:", d.get(search, "Not found"))
remove = input("Enter a key to remove: ")
d.pop(remove, None)
print("After removal:", d)

# --------------------------------------------------------------
# 9️⃣ CONDITIONALS
# --------------------------------------------------------------
print("\n--- Conditional Statements ---")
num = int(input("Enter a number to check if even/odd: "))
if num % 2 == 0:
    print(f"{num} is Even")
else:
    print(f"{num} is Odd")

# --------------------------------------------------------------
# 🔟 LOOPS
# --------------------------------------------------------------
print("\n--- Loops ---")
limit = int(input("Enter a number to loop up to: "))
print("For Loop:")
for i in range(1, limit + 1):
    print(i, end=" ")

print("\nWhile Loop:")
count = 1
while count <= limit:
    print(count, end=" ")
    count += 1

# --------------------------------------------------------------
# 1️⃣1️⃣ FUNCTIONS
# --------------------------------------------------------------
print("\n\n--- Functions ---")
def greet(name):
    return f"Hello, {name}! Welcome to the Python Demo."

user_name = input("Enter your name: ")
print(greet(user_name))

def add_numbers(x, y):
    return x + y

x = int(input("Enter first number for function: "))
y = int(input("Enter second number: "))
print("Sum using function:", add_numbers(x, y))

# --------------------------------------------------------------
# ✅ END OF PROGRAM
# --------------------------------------------------------------
print("\n🎯 End of Python Concepts Demo – You’ve just run a full mini-project!")
```
